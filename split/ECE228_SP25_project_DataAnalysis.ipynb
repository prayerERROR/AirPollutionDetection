```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
import os
```

```python
base_path = "/Users/majinbo/Desktop/228/Air Pollution Image Dataset/Air Pollution Image Dataset/Country_wise_Dataset/India"
nepal_path = "/Users/majinbo/Desktop/228/Air Pollution Image Dataset/Air Pollution Image Dataset/Country_wise_Dataset/Nepal/Biratnagar"

cities = ["Bengaluru", "Delhi", "Dimapur", "Faridabad", "Greater_Noida", "Mumbai", "Tamil_Nadu"]
pollutants = ['PM2.5', 'PM10', 'O3', 'CO', 'SO2', 'NO2']

pollutant_means_cities = {}

for city in cities:
    if city in ["Greater_Noida", "Mumbai"]:
        file_name = f"{city}_AQI_All.csv"
    else:
        file_name = f"{city}_AQI_All_Info.csv"
    
    city_file = os.path.join(base_path, city, file_name)
    df = pd.read_csv(city_file)
    pollutant_means_cities[city.replace("_", " ")] = df[pollutants].mean()


pollutant_means_df = pd.DataFrame(pollutant_means_cities).T
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
pollutant_means_df.plot(kind='bar', stacked=True, colormap='Set3', ax=ax)
plt.title('Stacked Bar Chart: Pollutant Composition in Indian Cities', fontsize=18, fontweight='bold')
plt.ylabel('Average Concentration')
plt.xlabel('City')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Pollutants')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(pollutant_means_df, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title('Heatmap: Pollutant Levels Across Indian Cities', fontsize=18, fontweight='bold')
plt.ylabel('City')
plt.xlabel('Pollutants')
plt.tight_layout()
```

```python
angles = [n / float(len(pollutants)) * 2 * pi for n in range(len(pollutants))]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
for city in pollutant_means_df.index:
    values = pollutant_means_df.loc[city].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=city)
    ax.fill(angles, values, alpha=0.1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(pollutants)
plt.title('Radar Chart: Pollutant Composition in Indian Cities', fontsize=18, fontweight='bold')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

```python

birat_file = os.path.join(nepal_path, "Biratnagar_AQI_All_info.csv")
df_birat = pd.read_csv(birat_file)
avg_birat = df_birat[pollutants].mean()


plt.figure(figsize=(8, 8))
plt.pie(avg_birat, labels=avg_birat.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title("Average Pollutant Composition in Biratnagar", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
